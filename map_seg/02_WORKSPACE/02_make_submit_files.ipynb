## Import

```python
import numpy as np
import pandas as pd
import cv2
import os
```

## Convert masks to RLE

```python
def mask_to_coordinates(mask):
    flatten_mask = mask.flatten()
    if flatten_mask.max() == 0:
        return f'0 {len(flatten_mask)}'
    idx = np.where(flatten_mask!=0)[0]
    steps = idx[1:]-idx[:-1]
    new_coord = []
    step_idx = np.where(np.array(steps)!=1)[0]
    start = np.append(idx[0], idx[step_idx+1])
    end = np.append(idx[step_idx], idx[-1])
    length = end - start + 1
    for i in range(len(start)):
        new_coord.append(start[i])
        new_coord.append(length[i])
    new_coord_str = ' '.join(map(str, new_coord))
    return new_coord_str
```

```python
maskspath = '/workspace/Competition/landmap/01_DATA/final_train/masks'
```

```python
masks = os.listdir(maskspath)
```

```python
masks
```

```python
test_maskpath = '/workspace/Competition/landmap/01_DATA/final_test/masks'
test_masks = os.listdir(test_maskpath)
```

```python
recid = []
rles = []

for mname in test_masks:
    maskpath = os.path.join(test_maskpath, mname)
    mask = cv2.imread(maskpath, cv2.IMREAD_GRAYSCALE)
    recid.append(mname)
    rle = mask_to_coordinates(mask)
    rles.append(rle)
```

## Make Answer File

Public:Private == 3:7

```python
isPublic = []
for i in range(len(recid)):
    samp = np.random.uniform(0,1)
    if samp < 0.3:
        isPublic.append(True)
    else:
        isPublic.append(False)
```

```python
isPublic
```

```python
classes = list(np.repeat('building',1303))
```

```python
answerdf = pd.DataFrame(list(zip(recid,classes,rles,isPublic)),columns=['img_id','class','prediction','public'])
```

```python
answerdf.to_csv('answer.csv',index=False)
```
